```python
# Import dependencies
import pandas as pd
import numpy as np
import requests
from config import weather_api_key
from citipy import citipy
from datetime import datetime
```

```python
# Generate 2000 random latitudes and longitudes
lats = np.random.uniform(-90,90,size=2000)
longs = np.random.uniform(-180,180,size = 2000)
tuplepairs = zip(lats,longs)
coordinates = list(tuplepairs)
```

```python
# Grab the nearest cities 
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
print(len(cities))

test = cities[:15]
test
```

```python
# Using the weather API, record the max temp, humidity, cloudiness, wind speed, weather description
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
test_url = url + "&q=" + "Boston"
test_data = requests.get(test_url).json()
test_data['weather'][0]['description']
```

```python
# Create counters.
record_count = 1
set_count = 1

city_data = []

for index, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes.
    if (index % 50 == 0 and index >= 50):
        set_count += 1
        record_count = 1
        

    city_url = url + "&q=" + city.replace(" ","+")
    
    try:
        
        # Parse the data
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_desc = city_weather["weather"][0]["description"]
        
        # Convert to timestamp
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        city_data.append({
            "City":city.title(),
            "Lat":city_lat,
            "Lng":city_lng,
            "Max Temp":city_max_temp,
            "Humidity":city_humidity,
            "Cloudiness":city_clouds,
            "Wind":city_wind,
            "Country":city_country,
            "Current Description":city_desc
        })
        
    except:
        print(f"{city} not found. :( Skipping...")
        pass
    
# Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count.
    record_count += 1
    
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert to dataframe

city_df = pd.DataFrame(city_data)
```

```python
# Export to CSV

output_file = 'Weather_Database/WeatherPy_Database.csv'
city_df.to_csv(output_file, index_label="City_ID")
```
